```python
import pandas as pd

df = pd.read_csv('./movie_data_clean.csv', index_col='tmdb_id')
```

```python
df
```

```python
df.dtypes
```

```python
useless_cols = ['Unnamed: 0', 'id', 'imdb_id', 'overview', 'release_date', 'title', 
               'belongs_to_collection', 'production_companies', 'production_countries', 
                'spoken_languages']
```

```python
df = df.drop(useless_cols, axis=1)
```

```python
df.dtypes
```

```python
df.astype(bool).sum(axis=0)
```

It seems like most budget and revenue values are zero. We decide to discard those two.

```python
df = df.drop(['budget', 'revenue'], axis=1)
```

# With normalization, run this

```python
normalized_cols = ['popularity', 'runtime', 'vote_average', 'vote_count']
```

```python
df[normalized_cols] = df[normalized_cols].apply(lambda x: (x - x.mean()) / x.std())
```

```python
df
```

# With onehot, run this

```python
one_hot_columns = ['adult', 'original_language', 'genres']
```

```python
df_onehot = pd.get_dummies(df, columns=one_hot_columns)
```

```python
df_onehot.head()
```

```python
df_onehot.dtypes
```

```python
df_onehot.to_csv('movie_onehot_adult_original_lang_genre.csv')
```
